```python
import random

# Word list for the game
word_list = ['python', 'hangman', 'challenge', 'developer', 'computer', 'science', 'algorithm']
chosen_word = random.choice(word_list)
word_length = len(chosen_word)

# Initial game setup
display = ['_'] * word_length
lives = 6
guessed_letters = []

# Hangman ASCII art
stages = [
    """
      _______
     |/      |
     |      (_)
     |      \\|/
     |       |
     |      / \\
     |
    _|___
    """,
    """
      _______
     |/      |
     |      (_)
     |      \\|/
     |       |
     |      /
     |
    _|___
    """,
    """
      _______
     |/      |
     |      (_)
     |      \\|/
     |       |
     |
     |
    _|___
    """,
    """
      _______
     |/      |
     |      (_)
     |      \\|
     |       |
     |
     |
    _|___
    """,
    """
      _______
     |/      |
     |      (_)
     |       |
     |       |
     |
     |
    _|___
    """,
    """
      _______
     |/      |
     |      (_)
     |
     |
     |
     |
    _|___
    """,
    """
      _______
     |/      |
     |
     |
     |
     |
     |
    _|___
    """
]

# Game loop
print("üéÆ Welcome to Hangman!")

while lives > 0 and '_' in display:
    print("\n" + ' '.join(display))
    print(stages[6 - lives])
    guess = input("Guess a letter: ").lower()

    if not guess.isalpha() or len(guess) != 1:
        print("‚ùó Please enter a single valid letter.")
        continue

    if guess in guessed_letters:
        print(f"‚ö†Ô∏è You've already guessed '{guess}'. Try another.")
        continue
    else:
        guessed_letters.append(guess)

    if guess in chosen_word:
        for i in range(word_length):
            if chosen_word[i] == guess:
                display[i] = guess
        print(f"‚úÖ Good guess! '{guess}' is in the word.")
    else:
        lives -= 1
        print(f"‚ùå Sorry, '{guess}' is not in the word. You lost a life.")

# Game result
if '_' not in display:
    print("\nüéâ You won! The word was:", chosen_word)
else:
    print(stages[6])
    print("\nüíÄ Game Over! The word was:", chosen_word)
```
